```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# min support of 0.05
support = apriori(df, min_support=0.05, use_colnames = True)
print(len(support))
```

```python
# define main set of rules
rules = association_rules(support, metric='lift',min_threshold=1)
print(len(rules))
```

```python
# looking at top 20 survivor rules, sort by lift
show_cols = ['antecedents','consequents','support','confidence','lift']
survived_rules = rules[rules['consequents']=={'Survived'}]
survived_rules[show_cols].sort_values('lift', ascending=False).head(10)
```
